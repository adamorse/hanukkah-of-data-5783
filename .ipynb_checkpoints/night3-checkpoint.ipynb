### Year of the Dog

Lets start with accessing customer info. We'll need
- phone for the answer
- birthdate for Aries and Year of the Dog
- address and citystatezip for the neighborhood
- might as well include customerid and name just in case

```python
import pandas as pd
customers = pd.read_csv('noahs-customers.csv',usecols=['customerid','name','phone','birthdate','address','citystatezip'])
customers.head()
```

Let's split up year, month, and day from `birthdate`

```python
birth = customers['birthdate'].str.split('-',expand=True)
birth.head()
```

```python
customers['birthday'] = birth[2].astype(int)
customers['birthmonth'] = birth[1].astype(int)
customers['birthyear'] = birth[0].astype(int)
customers.head()
```

Let's filter for Year of the Dog birthdays.

Google says every 12 years, 2006 is one of those years

```python
dog_years = [2006 - 12*i for i in range(5)]
dog = customers[customers.birthyear.isin(dog_years)]
dog.head()
```

### Aries

Google says March 21st through April 19th

```python
is_aries = ((dog.birthmonth ==3) & (dog.birthday >= 21)) | ((dog.birthmonth ==4) & (dog.birthday <= 19))
aries = dog[is_aries]
aries.head()
```

### Neighborhood

Going back to `night1`, Jeremy Davis (the contractor) lived at 134-10 Foch Blvd	South Ozone Park, NY 11420	

```python
nbhd = aries[aries.citystatezip.str.contains('Ozone')]
nbhd.head()
```
